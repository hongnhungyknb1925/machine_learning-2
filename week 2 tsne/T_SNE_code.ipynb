# iris

```python
import pandas as pd 
import numpy as np 
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/hongnhungyknb1925/machine_learning-2/main/week%202%20tsne/iris.data",header= None)
df.head(n=3)
```

```python
X = df[4]
X.head(n= 3 )
```

```python
del df[4]
dt = df.to_numpy()
```

```python
df_new = TSNE(n_components = 2, learning_rate = 'auto', init = 'random').\
                                                        fit_transform(dt)
df_new.shape
```

```python
df_new = pd.DataFrame(data = df_new)
df_new = pd.concat([df_new,X],axis = 1)
df_new
```

```python
df_new[4].unique()
```

```python
yellow = df_new[df_new[4] == 'Iris-setosa'] 
red = df_new[df_new[4] == 'Iris-versicolor'] 
glu = df_new[df_new[4] == 'Iris-virginica']
```

```python
plt.scatter(yellow[0],yellow[1],c='yellow')
plt.scatter(red[0],red[1],c='r')
plt.scatter(glu[0],glu[1],c='g')
plt.show()
```

# mnist

```python
import numpy as np 
import pandas as pd 
from sklearn.manifold import TSNE
```

```python
import tensorflow as tf 
from tensorflow import keras 
mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test)= mnist.load_data()
```

```python
Y = pd.DataFrame(y_train)
Y.head()
print(Y.value_counts())
```

```python
x_train.shape
```

```python
X = np.zeros((x_train.shape[0], 784))
for i in range(x_train.shape[0]):
    X[i] = x_train[i].flatten()
X = pd.DataFrame(X)
```

```python
X.head(n= 2)
```

```python
df = X
```

```python
tsne = TSNE(n_components= 2, learning_rate  = 'auto', init = 'random')
tsne = tsne.fit_transform(df.values)
```

```python
df['label'] = Y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Create the figure
fig = plt.figure( figsize=(8,8) )
ax = fig.add_subplot(1, 1, 1, title='TSNE' )
# Create the scatter
ax.scatter(
    x=tsne[:,0], 
    y=tsne[:,1], 
    c=df['label'], 
    cmap=plt.cm.get_cmap('Paired'), 
    alpha=0.4)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca = pca.fit_transform(df.values)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Create the figure
fig = plt.figure( figsize=(8,8) )
ax = fig.add_subplot(1, 1, 1, title='TSNE' )
# Create the scatter
ax.scatter(
    x=pca[:,0], 
    y=pca[:,1], 
    c=df['label'], 
    cmap=plt.cm.get_cmap('Paired'), 
    alpha=0.4)
plt.show()
```
